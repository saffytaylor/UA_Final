### Fixing the Shapefile 

```python
import geopandas as gpd
import pandas as pd
```

```python
york_electoral_shape = gpd.read_file('data/York_boundaries/original_shp/england_ua_wa_2001.shp')
```

```python
york_electoral_shape.head()
```

```python
york_update = york_electoral_shape.rename(columns={"name": "geo_label"})
```

```python
geo_codes = pd.read_csv("data/York_boundaries/geo_codes.csv", low_memory=False)
```

```python
merge_york = york_update.merge(geo_codes, left_on='geo_label', right_on='geo_label', how='right')
```

```python
merge_york = merge_york.drop(['label'], axis=1)
```

```python
merge_york
```

```python
merge_york.to_file('data/York_boundaries/york_electoral_boundaries.shp')  
```
